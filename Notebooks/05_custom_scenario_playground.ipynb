# Custom scenario playground

Use this notebook to override preset parameters and explore how regime behavior changes.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

from trace.api import make_dense_n_grid, run_reference_scenario, run_all_reference_scenarios
from trace.plotting import make_scenario_figure, make_summary_grid

n_grid = make_dense_n_grid(
    n_min=30,
    n_mid=1200,
    n_max=5000,
    n_low=16,
    n_high=8,
)

scenario = "persistent_advantage"

overrides = dict(
    d=2600,
    p=8,
    r=650,
    lam=18.0,
    sigma_z=0.22,
    T_mode="identity_like",
    T_jitter=0.02,
    n_pair=30000,
    ridge_alpha=0.5,
    C_X=0.008,
    C_H=1.6,
)

ds, result = run_reference_scenario(
    regime=scenario,
    n_samples=4000,
    dataset_seed=7,
    diagnostic_seed=123,
    n_grid=n_grid,
    repeats=64,
    train_fraction=0.7,
    **overrides,
)
```

```python
print(json.dumps(ds["meta"], indent=2))
print(json.dumps(result["summary"], indent=2))
print(json.dumps(result["regime_scores"], indent=2))
```

```python
fig = make_scenario_figure(
    scenario_name=f"{scenario}_custom",
    result=result,
    meta=ds["meta"],
)
plt.show()
```
