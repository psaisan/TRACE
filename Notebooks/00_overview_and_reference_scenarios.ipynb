# TRACE overview and reference scenarios

This notebook runs the four didactic TRACE regimes and summarizes their finite-sample behavior.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

from trace.api import make_dense_n_grid, run_reference_scenario, run_all_reference_scenarios
from trace.plotting import make_scenario_figure, make_summary_grid

n_grid = make_dense_n_grid(
    n_min=30,
    n_mid=1200,
    n_max=5000,
    n_low=16,
    n_high=8,
)

scenarios = [
    "sample_efficiency",
    "persistent_advantage",
    "no_advantage",
    "lossy_translation",
]

all_results, all_meta = run_all_reference_scenarios(
    scenarios=scenarios,
    n_samples=4000,
    n_grid=n_grid,
    repeats=64,
    train_fraction=0.7,
)
```

```python
summary_rows = []
for scenario in scenarios:
    s = all_results[scenario]["summary"]
    summary_rows.append({
        "scenario": scenario,
        "classification": s["classification"],
        "low_label_arc": s["low_label_arc"],
        "tail_arc": s["tail_arc"],
        "peak_arc": s["peak_arc"],
        "final_gap": s["final_gap"],
        "best_n": s["best_n"],
        "crossover_n": s["crossover_n"],
    })

summary_df = pd.DataFrame(summary_rows)
summary_df
```

```python
fig = make_summary_grid(all_results=all_results)
plt.show()
```

## Interpretation guide

- **sample efficiency**: early positive ARC that shrinks toward zero.
- **persistent advantage**: ARC remains positive at the right tail.
- **no advantage**: ARC stays near zero throughout.
- **lossy translation**: ARC is negative over much or all of the range.
